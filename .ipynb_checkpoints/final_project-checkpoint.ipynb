```python
from utils.data import *
from time import time
import matplotlib.pyplot as plt
```

```python
t1 = time()
data, labels, label_to_breed, breed_to_label = load_data()
print('Time to load the data: %ss' % int(time()-t1))
```

```python
shape = {}
shape['height'] = np.mean([data[i].shape[0] for i in range(len(data))])
shape['width'] = np.mean([data[i].shape[1] for i in range(len(data))])
shape['depth'] = np.mean([data[i].shape[2] for i in range(len(data))])
shape
```

```python
#one image has a depth of 4
errors, name_errors = [d for d in data if d.shape[2] != 3], [label_to_breed[labels[i]] 
                                                             for i in range(len(data)) if data[i].shape[2] != 3]
print(errors[0].shape)
plt.imshow(errors[0])
plt.title(name_errors[0])
```

```python
#change size of images (I took the mean)
t1 = time()
data, labels, label_to_breed, breed_to_label = load_data(width = 442, height = 385)
print('Time to load the data: %ss' % int(time()-t1))
```

```python
shape = {}
shape['height'] = np.mean([data[i].shape[0] for i in range(len(data))])
shape['width'] = np.mean([data[i].shape[1] for i in range(len(data))])
shape['depth'] = np.mean([data[i].shape[2] for i in range(len(data))])
shape
```

```python
# Visualization of the original data
fig = plt.figure(figsize=(20,10))
mask=np.random.choice(len(data),16, replace = False)
for i in range(16):
    ax = fig.add_subplot(4,4,i+1)
    ax.imshow(data[mask[i]])
    ax.axis('off')
    ax.title.set_text(label_to_breed[labels[mask[i]]])
```
